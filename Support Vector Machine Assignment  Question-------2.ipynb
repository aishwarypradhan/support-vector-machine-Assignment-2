```python
# classify the Size_Categorie using SVM

# month	month of the year: 'jan' to 'dec'
# day	day of the week: 'mon' to 'sun'
# FFMC	FFMC index from the FWI system: 18.7 to 96.20
# DMC	DMC index from the FWI system: 1.1 to 291.3
# DC	DC index from the FWI system: 7.9 to 860.6
# ISI	ISI index from the FWI system: 0.0 to 56.10
# temp	temperature in Celsius degrees: 2.2 to 33.30
# RH	relative humidity in %: 15.0 to 100
# wind	wind speed in km/h: 0.40 to 9.40
# rain	outside rain in mm/m2 : 0.0 to 6.4
# Size_Categorie 	the burned area of the forest ( Small , Large)
```

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("D:/forestfires.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
##Dropping the month and day columns
df.drop(["month","day"],axis=1,inplace =True)
```

```python
df
```

```python
X = df.iloc[:,0:28]
Y = df.iloc[:,28]
```

# Transformation of Categorical and Numerical Variable..................

```python
from sklearn.preprocessing import LabelEncoder
LE= LabelEncoder()
Y1=LE.fit_transform(Y)
```

```python
Y1=pd.DataFrame(Y1)
```

```python
Y1.columns=["size_category"]
```

```python
Y1
```

```python
from sklearn.preprocessing import StandardScaler
SS=StandardScaler()
X1=SS.fit_transform(X)
X1=pd.DataFrame(X1)
```

```python
X1
```

```python
X1.columns=['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain', 'area',
       'dayfri', 'daymon', 'daysat', 'daysun', 'daythu', 'daytue', 'daywed',
       'monthapr', 'monthaug', 'monthdec', 'monthfeb', 'monthjan', 'monthjul',
       'monthjun', 'monthmar', 'monthmay', 'monthnov', 'monthoct', 'monthsep']
```

```python
X1
```

```python
# Splitting Train and Test
from sklearn.model_selection._split import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X1,Y1,test_size=0.3,random_state=5)
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='linear', C=1)
clf.fit(X_train, y_train)
```

```python
y_pred_train = clf.predict(X_train)
y_pred_test  = clf.predict(X_test)
```

```python
from sklearn import metrics
print("Training Accuracy :",metrics.accuracy_score(y_train, y_pred_train).round(2))
print("Testing Accuracy:",metrics.accuracy_score(y_test, y_pred_test).round(2))
```
